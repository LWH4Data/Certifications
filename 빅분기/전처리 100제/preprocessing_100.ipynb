# Question 52.
- neighbourhood 값과 neighbourhood_group 값에 따른 price의 평균을 계층적으로 indexing 없이 구하여라.

---

## < pandas groupby의 as_index와 unstack 상세 설명 >

### 1. groupby의 as_index 옵션

- `as_index=True` (기본값)
    - groupby에 사용된 컬럼이 **결과 DataFrame의 인덱스**(복합/계층 인덱스)로 들어감
- `as_index=False`
    - groupby에 사용된 컬럼이 **일반 컬럼**으로 결과 DataFrame에 포함됨

---

### 2. 예시 코드

```python
Ans = df.groupby(['neighbourhood', 'neighbourhood_group']).agg({'price': ['mean']})
print(Ans)
```

- 결과: `neighbourhood`, `neighbourhood_group`가 **복합(MultiIndex) 인덱스**로 작용

---

### 3. unstack()의 작동 원리

- 복합 인덱스 구조에서 **하나의 인덱스 레벨을 컬럼으로 올려** wide-form 테이블로 변환
- ex)  
    - 행 인덱스: (`neighbourhood`, `neighbourhood_group`)
    - `unstack()` 적용 시:  
        - `neighbourhood_group`이 컬럼이 되고  
        - `neighbourhood`만 행 인덱스로 남음

---

### 4. 전체 코드 예시

```python
Ans = df.groupby(['neighbourhood', 'neighbourhood_group']).agg({'price': ['mean']}).unstack()
print(Ans)
```

- 결과:
    - **행:** neighbourhood
    - **열:** (price, mean, neighbourhood_group)
    - **값:** price의 평균

---

### 5. as_index 옵션에 따른 차이

- `as_index=True` (기본값):  
    - groupby 컬럼이 인덱스가 됨 → unstack() 적용 가능
- `as_index=False`:  
    - groupby 컬럼이 일반 컬럼 → unstack()이 의미 있는 wide-form 변환을 할 수 없음

---

### 6. 요약

- `groupby` + `as_index=True` + `unstack()` 조합은  
    - 그룹별 집계 결과를 **피벗 테이블처럼 넓은 형태**로 보고 싶을 때 유용
- `as_index=False`에서는 주로 단순 집계/머지/내보내기 용도로 사용


````python
# groupby의 as_index인자는 default가 True이다.
#   -> 따라서 neighbourhood, neighbourhood_group이 복합 index로 작용한다.
#   -> unstack() 함수를 적용하면 복합 index를 분할하고, 하나의 인덱스(현재 neighbourhood_group를 컬럼으로 사용한다.

Ans = df.groupby(['neighbourhood', 'neighbourhood_group']).agg({'price' : ['mean']}).unstack()
Ans
````

    - Less than $40K : a
    - $40K - $60K : b
    - $60K - $80K : c
    - $80K - $120K : d
    - $120K +’ : e

````python
# apply 함수는 함수를 각 요소(행 혹은 열 등)에 적용해 주는 함수이다.

def changeCategory(x):
    if x == 'Unkown' :
        return 'N'
    elif x == 'Less than $40K':
        return 'a'
    elif x == '$40K - $60K' :
        return 'b'
    elif x == '$60K - $80K' :
        return 'c'
    elif x == '$80K - $120K' :
        return 'd'
    elif x == '$120K +' :
        return 'e'
    
# newIncome의 각 요소에 함수를 적용.
df.newIncome = df.Income_Category.apply(changeCategory)
df.newIncome
````

# Question 80.
- 시간이 연속적으로 존재하며 결측치가 없는지 확인하라.

````python
col = '(년-월-일:시)'

# 1. 결측치 우선 확인
if df[col].isnull().sum():
    Ans = False
    
else:
    # 2. 정렬: 차분을 통해 시간이 연속적인지 확인하기 위해.
    df_sorted = df.sort_values(col)

    # 3. 차분을 통해 시간의 연속성 확인
    #   - iloc[1 : ]을 통해 NaT을 포함하지 않기.
    #   -> 조건문을 통해 확인할 것이기 때문에 NaT이 포함되면 안됨.
    diff = df_sorted[col].diff().iloc[1 : ]
    
    # if 절을 통한 확인
    '''
    < pd.Timedelta() vs pd.DateOffset() >
    - pd.Timedelta()를 일반적으로 더 많이 사용한다.
        - pd.Timedelta()는 스칼라
        - 시리즈 등 범용적으로 사용할 수 있지만 pd.DateOffset()은 시리즈 외의 자료구조에서는 error가 발생할 수 있기 때문.
    '''
    # all(): 주어진 값들이 모두 참인지 확인, 즉 시리즈가 모두 True인지 확인.
    if (diff == pd.Timedelta(hours = 1)).all():
        Ans = True
    else:
        Ans = False

Ans
````
